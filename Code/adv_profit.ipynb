## Import libraries

```python
import pandas as pd
import numpy as np
```

## Read the data

```python
data = pd.read_csv("../Data/50_AdAgency.csv")
data.head()
```

## Prepare the data

```python
X = data.iloc[:, 0:-2]
Y = data.iloc[:, -1].values
```

## Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape
```

## Normalize the data (Scale)

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
```

```python
X_train_standarded = standard_scaler.fit_transform(X_train)
X_test_transformed = standard_scaler.transform(X_test)
```

## Import and train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train_standarded, y_train)
```

```python
y_preds = linear_regression_model.predict(X_test_transformed)
y_preds
```

## Model's score

```python
score = linear_regression_model.score(X_test_transformed, y_test)
print(f"Multiple Linear Regression Model's Score is : {score:.2f}")
```
